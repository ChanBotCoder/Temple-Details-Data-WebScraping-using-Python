```python
import urllib.request
from inscriptis import get_text
import re
import pandas as pd
from csv import writer
import matplotlib.pyplot as plt
import numpy as np
```

```python
def txt_url(li,):
  url = str(li)
  html = urllib.request.urlopen(url).read().decode('utf-8')
  text = get_text(html)
  #print("The Website Content is:\n",text)
  with open('temp_prepar.txt', 'w') as f:
    f.write(text)
  print("Data Mining is Successfully")

  mylines = []                           
  with open ('temp_prepar.txt', 'rt') as myfile: 
    for myline in myfile:
      mylines.append(myline)
  print("Data processing is successfully")
  
  temp = []
  
  for i in range(len(mylines)):
    #mylines[i].replace(' ','\n')
    mylines[i] = mylines[i].strip()
    mylines[i] = re.sub(r'^\s+|\s+$', '', mylines[i])
    mylines[i] = re.sub(r'\s+', '-', mylines[i])
    mylines[i] = re.sub(r'^', '', mylines[i])
    mylines[i] = re.sub(r'$', '',mylines[i])
    #print(mylines[i])
    
  result_2 = [item.split('-') for item in mylines]
  
  result_3 = [value for value in result_2 if value != ['']]
  
  st = int(result_3.index(['Private', 'Temples'])+1)
  ed = int(result_3.index(['\ufeff']))
  
  midst = result_3.index(['Singers:'])
  mided = result_3.index(['Special', 'Features:']) +1
  
  result_4 = []
  for i in range(st,ed):
    val = result_3[i]
    result_4.append(val)
  
  result_5 = []
  for i in range(midst,mided):
    val = result_3[i]
    result_5.append(val)

  for i in range (len(result_3)):
    if result_3[i][0] == 'Home':
      temp_n=result_3[i]

  temp_na = []
  st = temp_n.index('>>')+1
  ed = temp_n.index('>')
  res_1 = temp_n[st:ed]
  res_2 = temp_n[ed+1:]
  temp_na.append(' '.join(res_1))
  temp_na.append(' '.join(res_2))
  
  temple_detail =[]
  
  def store_res(res):
    if res == [['', '']]:
      temple_detail.append('NA')
    else:
      temple_detail.append(' '.join(res[0]))
      
  detail = [['Festival:'],["Temple's", 'Speciality:'],['Opening', 'Time:'],['Address:'],['Phone:'],['General', 'Information:']]
  
  for i in range(len(detail)-1):
    temp_1 = result_5.index(detail[i])+1
    temp_2 = result_5.index(detail[i+1])
    res = result_5[temp_1:temp_2]
    if temp_1 == temp_2:
      res =  [['', '']]
      store_res(res)
    else:
      store_res(res)
  
  temple_name = []
  n=':'
  for i in range(len(result_4)):
    temp = result_4[i].index(n) +1
    res = result_4[i][temp:]
    if res == ['', '']:
      temple_name.append('NA')
    else:
      temple_name.append(' '.join(res))

   
  temple_original = temp_na + temple_name + temple_detail
  
  for i in range(len(temple_original)):
    if temple_original[i] == ' ':
      temple_original[i] = 'NA'
    elif temple_original[i] == '':
      temple_original[i] = 'NA'
  #print(temple_original)

  if temple_original[1] == 'sri temple':
    print(i," is empty")
  else:
    with open('Final.csv', 'a') as f_object:
      writer_object = writer(f_object)
      writer_object.writerow(temple_original)
      f_object.close()
    with open('Final.txt', 'a') as f:
      f.write('\n')
      for word in temple_original:
        f.write(word)
      f.close()
      
```
